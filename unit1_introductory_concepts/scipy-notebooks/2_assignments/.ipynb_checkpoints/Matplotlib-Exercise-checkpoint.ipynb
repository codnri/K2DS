# Matplotlib
As it turns out, people are very terrible at understanding numerical data, but can process and interpret visual information at remarkable speeds -- quite the opposite of computers in fact, and as such, you will nearly always want some sort of visual to accompany your analysis.  In this exercise, we'll be using `Matplotlib`, a package in `SciPy`, utilizying MATLAB-like syntax, to generate many plots.

# Basics Using Generated Data
1. Generate three arrays of 500 values, $x$, $y_1$, $y_2$ such that $$\{x \mid -2\pi \le x \le 2\pi \}$$ $$y_1 = sin(x)$$ $$y_2 = cos(x)$$
1. Using the default settings, use pyplot to plot $y_1$ and $y_2$ versus $x$, all on the same plot.
1. Generate the same plots, but set the horizontal and vertical limits to be slightly smaller than the default settings.  In otherwords, tighten up the plot a bit.
1. Generate the same plots using all settings from above, but now change the color and thickness of each from the defaults.  Play around with the values a bit until you are satisfied with how they look.
1. Generate the same plots using all settings from above, but now add some custom tickmarks with labels of your choosing.  Which values would make sense given the functions we are using?
1. Generate the same plots using all the settings from above, but now change your plot spines so that they are centered at the origin.  In other words, change the plot area from a "box" to a "cross".
1. Generate the same plots using all the settings from above, but now add a legend, with labels `sine` and `cosine`, to your plot in a position of your choosing.
1. Now generate two more data sets, $$y_3 = sin(x) + sin(2x)$$ $$y_3 = cos(x) + cos(2x)$$ and add them to your plot, setting different color and line styles (for example, dotted).  Be sure to adjust your scales and legend as needed.  Also add a title to your plot.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2*np.pi,2*np.pi,500,endpoint=True)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# Generate the same plots, but set the horizontal and vertical limits 
# to be slightly smaller than the default settings. In otherwords, tighten up the plot a bit.
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(-6,6)
plt.ylim(-1,1)
plt.show()
```

```python
#4. Generate the same plots using all settings from above, but now change the color 
# and thickness of each from the defaults. Play around with the values a bit 
# until you are satisfied with how they look.
plt.plot(x,y1,color='red',linewidth=3.0)
plt.plot(x,y2,color='blue',linewidth=3.0)
plt.xlim(x.min(),x.max())
plt.ylim(min(y1.min(),y2.min()),max(y1.max(),y2.max()))
plt.show()

```

```python
#5. Generate the same plots using all settings from above,
# but now add some custom tickmarks with labels of your choosing.
# Which values would make sense given the functions we are using?
plt.plot(x,y1,color='red',linewidth=3.0)
plt.plot(x,y2,color='blue',linewidth=3.0)
plt.xlim(x.min(),x.max())
plt.ylim(min(y1.min(),y2.min()),max(y1.max(),y2.max()))
plt.xticks(np.linspace(-2*np.pi,2*np.pi,9),
           [r'$-2\pi$',r'$-(2/3)\pi$',r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$',r'$+(2/3)\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])
plt.show()

```

```python
#6. Generate the same plots using all the settings from above, but now change your plot spines 
# so that they are centered at the origin. 
# In other words, change the plot area from a "box" to a "cross".
plt.plot(x,y1,color='red',linewidth=3.0)
plt.plot(x,y2,color='blue',linewidth=3.0)
plt.xlim(x.min(),x.max())
plt.ylim(min(y1.min(),y2.min()),max(y1.max(),y2.max()))
plt.xticks(np.linspace(-2*np.pi,2*np.pi,9),
           [r'$-2\pi$',r'$-(2/3)\pi$',r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$',r'$+(2/3)\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])

ax = plt.gca()
ax.spines['right'].set_color('none') # remove the right spines
ax.spines['top'].set_color('none') # remove the top spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
plt.show()
```

```python
# Generate the same plots using all the settings from above,
# but now add a legend, with labels sine and cosine,
# to your plot in a position of your choosing.
scale = 1.7
plt.plot(x,y1,color='red',linewidth=3.0,label='sine') #★
plt.plot(x,y2,color='blue',linewidth=3.0,label='cosine') #★
# plt.xlim(scale*x.min(), scale*x.max())
plt.xlim(6.4, 6.4)
plt.ylim(scale*min(y1.min(),y2.min()), scale*max(y1.max(),y2.max()))
plt.xticks(np.linspace(-2*np.pi,2*np.pi,9),
           [r'$-2\pi$',r'$-(2/3)\pi$',r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$',r'$+(2/3)\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])

ax = plt.gca()
ax.spines['right'].set_color('none') # remove the right spines
ax.spines['top'].set_color('none') # remove the top spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc='upper left', frameon=False) #★
plt.show()
```

```python
#8. Now generate two more data sets,
# y3=sin(x)+sin(2x)
# y3=sin(x)+sin(2x)
# and add them to your plot, setting different color and line styles (for example, dotted).
# Be sure to adjust your scales and legend as needed. Also add a title to your plot.

scale = 2.1
y3=np.sin(x)+np.sin(2*x)
y4=np.cos(x)+np.cos(2*x)
plt.plot(x,y1,color='red',linewidth=3.0,label='sin(x)') 
plt.plot(x,y2,color='blue',linewidth=3.0,label='cos(x)') 
plt.plot(x,y3,color='orange',linewidth=2.5,linestyle=':',label='sin(x)+sin(2x)') #★
plt.plot(x,y4,color='purple',linewidth=2.5,linestyle=':',label='cos(x)+cos(2x)') #★

# plt.xlim(scale*x.min(), scale*x.max())
plt.xlim(6.4, 6.4)
# plt.ylim(scale*min(y1.min(),y2.min()), scale*max(y1.max(),y2.max()))
plt.ylim(-4.0,4.0)
plt.xticks(np.linspace(-2*np.pi,2*np.pi,9),
           [r'$-2\pi$',r'$-(2/3)\pi$',r'$-\pi$',r'$-\pi/2$',r'$0$',r'$+\pi/2$',r'$+\pi$',r'$+(2/3)\pi$',r'$+2\pi$'])
plt.yticks([-1,0,1],[r'$-1$',r'$0$',r'$+1$'])

ax = plt.gca()
ax.spines['right'].set_color('none') # remove the right spines
ax.spines['top'].set_color('none') # remove the top spines
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

plt.legend(loc='upper left', frameon=False,prop={'size':8}) #★
plt.title('Waves')
plt.show()
```

Clean it up and merge it with your original aircraft accident data above.  Call this merged data set `data_all`.

```python
import pandas as pd
data_acc = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/gamclass/airAccs.csv',index_col=0)
data_acc.head()
```

```python
# 2. Generate a histogram for the number of deaths, using bin sizes of your choice.
# Be sure to adjust the axis and to add a title to make your plot aesthetically appealing.
plt.hist(data_acc.Dead.dropna(),bins=50)
plt.axis([0,165,0,3300])
plt.xticks(np.arange(0,180,20))
plt.title('Death in an AirLine Accident')
plt.ylabel('Count')
plt.xlabel('Deaths')
plt.show()
```

```python
#3. Make some plots of total number of deaths with respect to time,
# making use of Pandas time series functionality. 
# Again, be sure to make your plot aesthetically appealing.
ts = data_acc.copy()
ts = ts.set_index(pd.DatetimeIndex(ts['Date'])).drop('Date',axis=1) # set 'Date' col as index and then, remove it

t = pd.date_range(start=data_acc['Date'][1],end=data_acc['Date'][len(data_acc)])
# print(ts.loc[pd.DatetimeIndex(['1913-09-09','1908-09-18'])])
ts = ts.loc[t]
ts = ts.fillna(0)
ts['year'] = ts.index.year
ts_yearly_tot = ts.groupby('year').sum()['Dead']
# ts_yearly_tot.head()
# ts.groupby('year').sum()
```

```python
# ts2 = data_acc.copy()
# pd.DatetimeIndex(ts2['Date'])
# ts2.drop('Date',axis=1).head()

plt.plot(ts_yearly_tot,linewidth=2.5)
plt.xlim(ts_yearly_tot.index[0],ts_yearly_tot.index[len(ts_yearly_tot.index)-1])
plt.xlabel('Year')
plt.ylabel('Total Deaths') 
plt.title('Total Deaths per Year')
plt.show()
```

```python
#4. We're now going to add in some data from a different source to take a look at the bigger picture 
# in terms of number of passengers flying each year.
# Head over to the World Bank Webpage and download the .csv version of the data in the link.
# Clean it up and merge it with your original aircraft accident data above. Call this merged data set data_all
data_pass_raw = pd.read_csv('./data/pass_data.csv', encoding='ISO-8859-1')
data_pass_raw.head()
```

```python
data_pass_raw.dtypes.head()
```

```python
data_pass = data_pass_raw.iloc[:,4:].sum() # pick up columns after 4th
data_pass = pd.DataFrame({'year':data_pass.index.astype(float),'Passengers':data_pass})
data_acc_temp = data_acc[['Date','Dead']]
data_acc_temp = data_acc_temp.set_index(pd.DatetimeIndex(data_acc['Date']))
data_acc_temp['year']=data_acc_temp.index.year
data_acc_temp = data_acc_temp.groupby('year').sum()
data_acc_temp = pd.DataFrame({'year':data_acc_temp.index.astype(float),'Dead':data_acc_temp.Dead})
data_all = pd.merge(data_pass,data_acc_temp)[['year', 'Passengers', 'Dead']]
data_all = data_all[ (data_all.Passengers != 0) & (data_all.Dead != 0) ]
data_all.head()

```

```python

#5. Using data_all, create two graphs to visualize how the number of deaths and passengers vary with time,
# and, as always, make your plots as visually appealing as possible.
data_all['per_M'] = (data_all.Dead/data_all.Passengers)* 10000000
data_all['year'] = [int(y) for y in data_all.year]

plt.figure(1)
plt.subplot(3,1,1)
plt.plot(data_all.year, data_all.Passengers / 1000000)

plt.figure(2)
plt.subplot(3,1,2)
plt.plot(data_all.year, data_all.Dead)
plt.ylabel('Deaths')


plt.figure(3)
plt.subplot(3,1,3)
plt.plot(data_all.year, data_all.per_M)
plt.ylabel('Deaths per Million')

plt.show()
```

```python

data_all['per_M'] = (data_all.Dead/data_all.Passengers)* 10000000
data_all['year'] = [int(y) for y in data_all.year]

plt.figure(1)
plt.subplot(3,1,1)
plt.bar(data_all.year, data_all.Passengers / 1000000)

plt.figure(2)
plt.subplot(3,1,2)
plt.bar(data_all.year, data_all.Dead)
plt.ylabel('Deaths')


plt.figure(3)
plt.subplot(3,1,3)
plt.bar(data_all.year, data_all.per_M)
plt.ylabel('Deaths per Million')

plt.show()
```

```python
#6. Make a pie chart representing the number of deaths for each decade.
# Consult the pyplot documentation to play around with the settings a bit.
data_all_d = data_all
data_all_d['dec'] = data_all_d.year // 10 * 10
data_all_d = data_all_d.groupby('dec').sum()[['Dead']]

labels = '1970', '1980', '1990', '2000', '2010'
colors = ['blue', 'red', 'yellow', 'green', 'magenta']
plt.pie(data_all_d.Dead, colors = colors, autopct = '%1.1f%%', shadow = True, startangle=140)
plt.title('Percent aircraft deaths by decade')
plt.legend(labels, loc = 'best')
plt.axis('equal')
plt.show();
```
